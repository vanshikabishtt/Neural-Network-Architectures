```python
df1=df.drop(columns=["ID","Name","Subtitle","Icon URL","Average User Rating","User Rating Count","Price","In-app Purchases","Description","Age Rating","Languages","Size","Primary Genre","Genres","Original Release Date","Current Version Release Date"])
```

```python
import re
```

```python
def remove_tag(raw_text):
    cleaned_text = re.sub(r'[:/]', '', raw_text)
    return cleaned_text
```

```python
df1['URL'] = df1['URL'].apply(remove_tag)
```

```python
df1.head()
```

```python
df1['Developer'] = df1['Developer'].apply(lambda x:x.lower())
```

```python
df1.Developer[0]
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
stop_words = set(stopwords.words("english"))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
def work(text):
    # 1. Tokenization
    tokens = word_tokenize(text.lower())

    # 2. Stopword Removal 
    filtered = [w for w in tokens if w.isalpha() and w not in stop_words]

    # 3. Stemming
    stemming = [stemmer.stem(w) for w in filtered]

    # 4. Lemmatization
    lemmatization = [lemmatizer.lemmatize(w) for w in filtered]

    return tokens, filtered, stemming, lemmatization
```

```python
df1["tokens"] = df1["Developer"].apply(lambda x: work(x)[0])
df1["stopwords"] = df1["Developer"].apply(lambda x: work(x)[1])
df1["stemming"] = df1["Developer"].apply(lambda x: work(x)[2])
df1["lemmatization"] = df1["Developer"].apply(lambda x: work(x)[3])
```

```python
df1.head()
```

```python
def compare(a):
    vocab = set()
    for i in a:
        vocab.update(i)
    return len(vocab)
df1["tokens"] = df1["Developer"].apply(lambda x: work(x)[0])
df1["stopwords"] = df1["Developer"].apply(lambda x: work(x)[1])
df1["stemming"] = df1["Developer"].apply(lambda x: work(x)[2])
df1["lemmatization"] = df1["Developer"].apply(lambda x: work(x)[3])

answer = {
    "no of tokens": compare(df1["tokens"]),
    "after removal of stopwords": compare(df1["stopwords"]),
    "after stemming": compare(df1["stemming"]),
    "after lemmatization": compare(df1["lemmatization"])
}

pd.DataFrame(answer, index=["Comparision result"])
```
